```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("Adidas_US_Sales Datasets.xlsx")
```

```python
data
```

```python
# Droping the Unnamed and Formatting the table
data.columns = data.iloc[3]
data = data.iloc[4:,1:].reset_index(drop=True)
```

**Creating New Columns**

```python
# Creating New columns(Year, Month, Day)
data['Year'] = pd.to_datetime(data['Invoice Date']).dt.year
data['Month'] = pd.to_datetime(data['Invoice Date']).dt.month
data['Days'] = pd.to_datetime(data['Invoice Date']).dt.day
```

**Checking for Null Values:**

```python
data.isnull().sum()
```

Clean data from string to float

```python
data['Price per Unit'] = data['Price per Unit'].astype(float)
data['Total Sales'] = data['Total Sales'].astype(float)
data['Units Sold'] = data['Units Sold'].astype(float)
data['Operating Profit'] = data['Operating Profit'].astype(float)
```

```python
data[['Price per Unit','Total Sales','Units Sold','Operating Profit']].head(5)
```

**Maximum and Minimum amount**

```python
print('---Units sold---')
max = data['Units Sold'].max()
min = data['Units Sold'].min()
print(f'Maximum :', max)
print(f'Minimum :', min)

print('---Operating Profit---')
max = data['Operating Profit'].max()
min = data['Operating Profit'].min()
print(f'Maximum :', max)
print(f'Minimum :', min)

print('---Total Sales---')
max = data['Total Sales'].max()
min = data['Total Sales'].min()
print(f'Maximum :', max)
print(f'Minimum :', min)

print('---Price per Unit---')
max = data['Price per Unit'].max()
min = data['Price per Unit'].min()
print(f'Maximum :', max)
print(f'Minimum :', min)

print('---Operating Margin---')
max = data['Operating Margin'].max()*100
min = data['Operating Margin'].min()*100
print(f'Maximum :', max)
print(f'Minimum :', min)
```

**Total Units Sold, Profit Sales, Operating Profit**

```python
Total = data['Units Sold'].sum()
print(f'Total Units Sold: ', Total)

Total = data['Operating Profit'].sum()
print(f'Operating Profit: ', Total)

Total = data['Total Sales'].sum()
print(f'Total Sales: ', Total)
```

**Adding the Year column for the product**

```python
def gender(x):
  if 'Men' in x:
    return 'Men'
  elif 'Women' in x:
    return 'Women'

data['Gender'] = data['Product'].apply(gender)
```

**Total Sales and the Operating Profit based on the Gender of the Product**

**Adding column type of Product**

```python
def type(x):
  if 'Footwear' in x:
    return 'Footwear'
  elif 'Apparel' in x:
    return 'Apparel'

data['Type'] = data['Product'].apply(type)
```

**Total Units of Apparel and Footwear sold based on Gender**

```python
type = data.groupby(['Type','Gender'])[['Units Sold','Total Sales','Operating Profit']].sum().reset_index()
apparel = type[type['Type']=='Apparel']
footwear = type[type['Type']=='Footwear']
```

```python
apparel
```

```python
plt.figure(figsize=(8,5))
m = np.arange(len(apparel))
plt.xticks(m,apparel['Gender'])
plt.bar(m,footwear['Units Sold'], label='Footwear',width=0.3)
plt.bar(m+0.4,apparel['Units Sold'], label='Apparel',width=0.3)
plt.legend()
plt.xlabel('Gender')
plt.ylabel('Unit Sold')
plt.show()
```

**Total Sales and Profit of Apparel and Footwear based on Gender**

```python
plt.figure(figsize=(8,5))
m = np.arange(len(footwear))
plt.xticks(m,footwear['Gender'])

plt.bar(m, footwear['Total Sales'], label='Footwear Total Sales', width=0.3)
plt.bar(m+0.4, apparel['Total Sales'], label='Apparel Total Sales', width=0.3)

plt.bar(m, footwear['Operating Profit'], label='Footwear Profit', width=0.3)
plt.bar(m+0.4, apparel['Operating Profit'], label='Apparel Profit', width=0.3)

plt.legend()
plt.show()
```

**Top 20 City with high Total sales**

**Region with Lowest and Highest Total Sales**

```python
item = data[['Region', 'Total Sales']]
item = item.groupby('Region')[['Total Sales']].sum().reset_index()

lowest_region = item.loc[item['Total Sales'].idxmin()]
highest_region = item.loc[item['Total Sales'].idxmax()]

print(f"The Region with the lowest total sales is {lowest_region['Region']} with {lowest_region['Total Sales']} total sales")
print(f"The Region with the highest total sales is {highest_region['Region']} with {highest_region['Total Sales']} total sales")

sns.barplot(data=item, x=item['Region'], y = item['Total Sales'])
plt.show()
```

**Region with Lowest and Highest Operating Profit**

```python
item = data[['Region', 'Operating Profit']]
item = item.groupby('Region')[['Operating Profit']].sum().reset_index()

lowest_region = item.loc[item['Operating Profit'].idxmin()]
highest_region = item.loc[item['Operating Profit'].idxmax()]

print(f"The Region with the lowest Operating Profit is {lowest_region['Region']} with {lowest_region['Operating Profit']} profit")
print(f"The Region with the highest Operating Profit is {highest_region['Region']} with {highest_region['Operating Profit']} profit")

sns.barplot(data=item, x=item['Region'], y = item['Operating Profit'])
plt.show()
```

**Top 10 States/City by Operationg Profit**

```python
top_State_city = data.groupby(['State','City'])['Operating Profit'].sum().sort_values(ascending=False)
top_State_city.nlargest(10).plot(kind='bar', color='green')
plt.title('Top 10 State/Cities by Operating Profit')
plt.xlabel('State/Cities')
plt.ylabel('Profit')
plt.show()
```

**Correlation between 'Price Per Unit', 'Total Sales', 'Unit Sold' , 'Operating Profit'**

```python
corr = data[['Price per Unit','Total Sales','Units Sold','Operating Profit']].corr()
sns.heatmap(corr, cmap='coolwarm', annot= True, fmt= '.2f')
plt.xlabel(' ')
plt.ylabel(' ')
plt.title('Correlation Matrix')
plt.yticks(rotation = 0)
plt.show()
```
